```python
import numpy as np
import matplotlib.pyplot as plt

def convolve_through_aperture(input_image, aperture_image):
    # Take Fourier transforms of input and aperture images
    input_ft = np.fft.fft2(input_image)
    aperture_ft = np.fft.fft2(aperture_image)
    
    # Multiply the Fourier transforms element-wise
    convolved_ft = input_ft * aperture_ft
    
    # Take the inverse Fourier transform to obtain the convolved image
    convolved_image = np.fft.ifft2(convolved_ft).real
    
    # Plot input image and its Fourier transform
    plt.subplot(3, 3, 1)
    plt.imshow(input_image, cmap='gray')
    plt.title('Input image')
    
    plt.subplot(3, 3, 2)
    plt.imshow(np.abs(input_ft), cmap='gray')
    plt.title('Magnitude of Fourier transform')
    
    plt.subplot(3, 3, 3)
    plt.imshow(np.angle(input_ft), cmap='gray')
    plt.title('Phase of Fourier transform')
    
    # Plot output image and its Fourier transform
    plt.subplot(3, 3, 4)
    plt.imshow(convolved_image, cmap='gray')
    plt.title('Output image')
    
    plt.subplot(3, 3, 5)
    plt.imshow(np.abs(convolved_ft), cmap='gray')
    plt.title('Magnitude of Fourier transform')
    
    plt.subplot(3, 3, 6)
    plt.imshow(np.angle(convolved_ft), cmap='gray')
    plt.title('Phase of Fourier transform')
    
    # Plot convolved image and its Fourier transform
    plt.subplot(3, 3, 7)
    plt.imshow(convolved_image, cmap='gray')
    plt.title('Convolved image')
    
    plt.subplot(3, 3, 8)
    plt.imshow(np.abs(convolved_ft), cmap='gray')
    plt.title('Magnitude of Fourier transform')
    
    plt.subplot(3, 3, 9)
    plt.imshow(np.angle(convolved_ft), cmap='gray')
    plt.title('Phase of Fourier transform')
    
    # Show the plots
    plt.show()

    
    
input_image = plt.imread('input_image.jpg')
aperture_image = plt.imread('aperture_image.jpg')
convolve_through_aperture(input_image, aperture_image)
```
