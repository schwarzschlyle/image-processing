# Color Image

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the centers of the circles
center1 = (0, 0)
center2 = (7, 0)

# Define the angles for the circles
theta = np.linspace(0, 2*np.pi, 1000)

# Define the radii for the circles
radius1 = 3
radius2 = 2

# Draw the circles
x1 = radius1*np.cos(theta) + center1[0]
y1 = radius1*np.sin(theta) + center1[1]
plt.plot(x1, y1, 'black', linewidth=1)

x2 = radius2*np.cos(theta) + center2[0]
y2 = radius2*np.sin(theta) + center2[1]
plt.plot(x2, y2, 'black', linewidth=1)

# Fill the donuts with color
circle1 = plt.Circle(center1, radius1, color='blue')
plt.gca().add_patch(circle1)

circle2 = plt.Circle(center2, radius1-radius2, color='red')
plt.gca().add_patch(circle2)

# Display the plot
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to plot a circle
def plot_circle(center_x, center_y, radius, color):
    theta = np.linspace(0, 2 * np.pi, 1000)
    x = radius * np.cos(theta) + center_x
    y = radius * np.sin(theta) + center_y
    plt.plot(x, y, color=color, linewidth=50)

# Define the colors and radii for each circle
colors = ['blue', 'black', 'red', 'yellow', 'green']
radii = [0.6, 0.8, 1, 1.2, 1.4]

# Set the size of the plot and turn off axis ticks
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xticks([])
ax.set_yticks([])

# Plot each circle using the defined colors and radii
for color, radius in zip(colors, radii):
    plot_circle(0, 0, radius, color)
    
for color, radius in zip(colors, radii):
    plot_circle(10, 0, radius, color)

# Add a title and display the plot
plt.title('Olympics Logo')
plt.show()
```
